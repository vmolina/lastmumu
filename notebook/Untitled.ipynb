```python
%matplotlib inline
import time

import numpy as np
import scipy as sp
import tables
import matplotlib.pylab as plt
import seaborn as sns
```

```python
clusters = 10
p = 400

sigma_high = np.linalg.inv(sp.stats.wishart.rvs(p,np.eye(p)))
sns.clustermap(sigma_high)
plt.show()
sigma = np.linalg.inv(sp.stats.wishart.rvs(p, np.eye(p)))
sns.clustermap(sigma)
plt.show()
indices = np.arange(p)
np.random.shuffle(indices)
cl_size = int(round(p/clusters))
for cl in range(clusters):
    cl_indices = indices[cl*cl_size : min((cl+1)*cl_size,p)]
    print cl_indices
    sigma[cl_indices,cl_indices] = sigma_high[cl_indices,cl_indices]
x = np.random.multivariate_normal(np.zeros((p)),sigma, size=(200))
sns.clustermap(sigma)
plt.show()
cor = np.corrcoef(x.T)
sns.clustermap(cor)
plt.show()
cor.shape
```
